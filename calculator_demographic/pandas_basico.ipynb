## Inicio
Este proyecto tiene la finalidad de entender y practicar conceptos iniciales para el aprendizaje de data science , uno de estos es el manejo de librerias como pandas y matplotlib, asi como el uso de jupyter notebook para la creacion de reportes y analisis de datos.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Teniendo las librerias ya instaladas , comenzaremos leyendo datos , en este caso los datos se encontraran en un csv llamado "sale_data" , lo leeremos y almacenaremos en una variable llamada "sales" ,y para mostrarla usaremos el metodo head()

```python
sales = pd.read_csv('sales_data.csv',parse_dates=['Date'])
# parse_data sirve para convertir la columna de fecha en un objeto de fecha
sales.head()
```

Teniendo los datos guardados en una variable comenzaremos aprendiendo algunos comandos para manipularlos , como el metodo mean() que nos permite obtener la media de una columna

```python
sales['Customer_Age'].mean()
```

 otro metodo que usarmeos es plot , los cuales nos permite hacer una grafica con los datos que querramos , en este caso generaremos una grafica de densidad , ingresaremos el parametro kde para que nos muestre la densidad de los datos 

```python
sales['Customer_Age'].plot(kind='kde',figsize=(14,6))
```

Otra manera de ver la dispersion de los datos es poniendo como parametro kind = "box" , esto nos mostrara una grafica de dispersion , como una caja 

```python
sales['Customer_Age'].plot(kind='box',vert=False,figsize=(14,6))
```

Provemos para otros datos, por ejemplo la columna 'Order_Quantity' .

```python
sales['Order_Quantity'].mean()
```

```python
sales['Order_Quantity'].plot(kind='hist',bins=30,figsize=(14,6))
```

```python
sales['Order_Quantity'].plot(kind='box',vert=False,figsize=(14,6))
```

Conozcamos otros metodos de pandas , ahora veremos el metodo value_counts() , este nos permite contar los valores de una columna 

```python
sales['Year'].value_counts()
```

Usando los anterior podemos generar una grafica de pie , para esto usaremos el metodo plot.pie() 

```python
sales['Year'].value_counts().plot(kind='pie',figsize=(6,6))
```

usemos el metodo value_counts()  pero para otros datos 

```python
sales['Month'].value_counts()
```

```python
sales['Month'].value_counts().plot(kind='bar',figsize=(14,6))
```
